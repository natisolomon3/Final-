```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files into DataFrames
lebronReg_df = pd.read_csv('Lebron.csv')
lebronPlayoffs_df = pd.read_csv('LebronPlayoffs.csv')

# Calculate the average of each metric for regular season and playoffs
avg_lebronReg = lebronReg_df.mean()
avg_lebronPlayoffs = lebronPlayoffs_df.mean()

# Combine the averages into a single DataFrame for comparison
comparison_df = pd.DataFrame({
    'Regular Season': avg_lebronReg,
    'Playoffs': avg_lebronPlayoffs
})

# Plotting the bar graph for comparison
comparison_df.plot(kind='bar', color=['red', 'blue'])
plt.title('Average Metrics Comparison: Regular Season vs Playoffs')
plt.xlabel('Metrics')
plt.ylabel('Average Value')
plt.legend(title='Career')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV files
lebronReg_df = pd.read_csv('Lebron.csv')
lebronPlayoffs_df = pd.read_csv('LebronPlayoffs.csv')

# Set 'Season' as the index
lebronReg_df.set_index('PTS')
lebronPlayoffs_df.set_index('PTS')

# Plot Regular Season Data
lebronReg_df.plot(kind='bar')
plt.title('Regular Season Metrics Comparison')
plt.xlabel('Season')
plt.ylabel('Metrics')
plt.show()

# Plot Playoffs Data
lebronPlayoffs_df.plot(kind='bar')
plt.title('Playoffs Metrics Comparison')
plt.xlabel('Season')
plt.ylabel('Metrics')
plt.show()
```
